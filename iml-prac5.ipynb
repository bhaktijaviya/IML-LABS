```python
'''Write a NumPy program to implement following operation
# to find the maximum and minimum value of a given flattened array'''

import numpy as np

arr = np.arange(9).reshape((3,3))

print("flattened array:")
print(arr)

print("Max value of flattened array:", np.max(arr))
print("Min value of flattened array:", np.min(arr))
```

```python
''' to compute the mean, standard deviation, and variance of a given
 array along the second axis'''

import numpy as np

arr = np.array([[10, 20, 20],
                 [30, 35, 40],
                 [45, 60, 65]])


mean = np.mean(arr, axis=1)
standard = np.std(arr, axis=1)
variance = np.var(arr, axis=1)


print("Mean of axis:", mean)
print("Standard deviation of axis:", standard)
print("Variance of axis:", variance)
```

```python
''' Write a NumPy program to implement following operation
 to convert a list of numeric values into a one-dimensional NumPy'''

import numpy as np

list12 = [1, 2, 3, 4, 5, 6]
arr = np.array(list12)
print(arr)
```

```python
'''to create a 3*3 matrix with values into a one-dimensional Numpy array'''

import numpy as np

val = np.arange(1, 28)

mat = val.reshape(3, 9)

print(mat)
```

```python
'''to append the values at the end of an array'''

import numpy as np

arr = np.array([1, 2, 3])

print("ARRAY: ",arr)

arr2 = np.array([4, 5, 6])


appended = np.append(arr2, arr)

print("appended list: ",appended)
```

```python
'''to create another shape from an array without changing its data(3*2 to 2*3)'''

import numpy as np
original = np.array([[10, 15],
                     [20, 25],
                     [30, 35]])


reshaped = original.reshape(2, 3)

print("Original Array:")
print(original)

print("Reshaped Array:")
print(reshaped)
```

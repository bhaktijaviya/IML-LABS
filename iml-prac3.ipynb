```python
'''Write a NumPy program to implement following operation
 to split an array of 14 elements into 3 arrays, each with 2, 4, and 8 elements in the original order'''

import numpy as np

arr1 = np.arange(14)
print(arr1)
print("2 ELEMENT ARRAY: ", arr1[:2])
print("4 ELEMENT ARRAY: ", arr1[2:6])
print("8 ELEMENT ARRAY: ", arr1[6:14])
```

```python
'''to stack arrays horizontally (column wise)'''

import numpy as np

arr = np.arange(11)
print("VERTICAL ARRAY: ", arr)
horizontal_arr = arr.reshape(-1,1)
print("HORIZONTAL ARRAY")
print(horizontal_arr)
```
